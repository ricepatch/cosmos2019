```python
# calculate average day difference
dday = ((hdat1['day'][-1])-(hdat1['day'][0]))/(hdat1['day'].size)
print(dday)
```

```python
# print day column and find first day value
print(hdat1['day'])
```

```python
# create new array with evenly spaced day values
newday = np.arange(88)
newday = newday*dday
newday = newday+1350
print(newday)
```

```python
# creates newa array size newday with zeroes
newa = newday*0
newb = newday*0
newc = newday*0
newd = newday*0

newerra = newday*0
newerrb = newday*0
newerrc = newday*0
newerrd = newday*0
```

```python
# Loop to create a interpolated set of y values
for i in range(newday.size):
    mask = (hdat1['day']>newday[i]-5) & (hdat1['day']<newday[i]+5)
    # calculate temporary flux and error ranges
    tempfluxa = na[mask]
    temperra = nerra[mask]
    
    tempfluxb = nb[mask]
    temperrb = nerrb[mask]
    
    tempfluxc = nc[mask]
    temperrc = nerrc[mask]
    
    tempfluxd = nd[mask]
    temperrd = nerrd[mask]
    # take means and save as a new data point
    newa[i] = tempfluxa.mean()
    newerra[i] = ((temperra**2).mean())**0.5
    
    newb[i] = tempfluxb.mean()
    newerrb[i] = ((temperrb**2).mean())**0.5
    
    newc[i] = tempfluxc.mean()
    newerrc[i] = ((temperrc**2).mean())**0.5
    
    newd[i] = tempfluxd.mean()
    newerrd[i] = ((temperrd**2).mean())**0.5
```

```python
# loop the shift for some number of terms on each curve
chirange = input("Input a range of shifts: ")
chia = np.zeros(chirange)
chic = np.zeros(chirange)
chid = np.zeros(chirange)
for i in range(chirange):
    # "shift" set 1 left by 1 newday and compare it to base set
    # create new temporary comparison arrays for the sets to be compared and normalize
    tempb = newb[0:-(i+1)]
    tempb = tempb/np.mean(tempb)
    tempa = newa[(i+1):]
    tempa = tempa/np.mean(tempa)
    tempc = newc[(i+1):]
    tempc = tempc/np.mean(tempc)
    tempd = newd[(i+1):]
    tempd = tempd/np.mean(tempd)
    # do the same for errors
    prechierrb = newerrb[0:-(i+1)]
    prechierrb = prechierrb/np.mean(prechierrb)
    prechierra = newerra[(i+1):]
    prechierra = prechierra/np.mean(prechierra)
    prechierrc = newerrc[(i+1):]
    prechierrc = prechierrc/np.mean(prechierrc)
    prechierrd = newerrd[(i+1):]
    prechierrd = prechierrd/np.mean(prechierrd)
    # subtract these arrays and take the squares of the differences, plus error
    tempsquaresab = (tempa-tempb)**2/(prechierra**2+prechierrb**2)
    tempsquarescb = (tempc-tempb)**2/(prechierrc**2+prechierrb**2)
    tempsquaresdb = (tempd-tempb)**2/(prechierrd**2+prechierrb**2)
    # sum these squares
    tempchi2ab = tempsquaresab.sum()
    tempchi2cb = tempsquarescb.sum()
    tempchi2db = tempsquaresdb.sum()
    # add the sums to new arrays for each
    chia[i] = tempchi2ab
    chic[i] = tempchi2cb
    chid[i] = tempchi2db
```

```python
# print the index of ideal shift
print("the ideal number of shifts for set a is", np.argmin(chia))
print("the ideal number of shifts for set c is", np.argmin(chic))
print("the ideal number of shifts for set d is", np.argmin(chid))
# shift the graph that many times
plt.errorbar(newday-dday*np.argmin(chia), newa, yerr=newerra, fmt='C0o')
plt.errorbar(newday, newb, yerr=newerrb, fmt='C1o')
plt.errorbar(newday-dday*np.argmin(chic), newc, yerr=newerrc, fmt='C2o')
plt.errorbar(newday-dday*np.argmin(chid), newd, yerr=newerrd, fmt='C3o')
```
